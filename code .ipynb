```python
import matplotlib.pyplot as plt
import networkx as nx
# Create a directed graph
G = nx.DiGraph()

steps = [
    "Gene Expression Data Collection (GEO)",
    "Data Preprocessing & Normalization",
    "Differential Gene Expression Analysis",
    "Identification of Shared DEGs",
    "KEGG Pathway Enrichment Analysis",
    "Biological Interpretation"
]

# Add nodes and edges
for i in range(len(steps) - 1):
    G.add_edge(steps[i], steps[i+1])

# Draw the flowchart
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G, seed=42)
nx.draw(
    G, pos,
    with_labels=True,
    node_size=3000,
    node_color="lightgray",
    font_size=9,
    font_weight="bold",
    arrows=True
)
plt.title("Bioinformatics Workflow for SARS-CoV-2 and Alzheimer's Disease Analysis")
plt.show()
```

```python
# KEGG pathway enrichment results
pathways = ["Alzheimer's disease (hsa05022)", "SARS-CoV-2 infection (hsa05171)"]
adjusted_pvalues = [0.002, 0.008]

plt.figure(figsize=(6,4))
plt.bar(pathways, adjusted_pvalues)
plt.ylabel("Adjusted p-value")
plt.title("KEGG Pathway Enrichment Analysis")
plt.show()
```

```python
shared_genes = ["APP", "MAPT", "SNCA", "IL6", "TNF"]
counts = [1, 1, 1, 1, 1]

plt.figure(figsize=(6,4))
plt.bar(shared_genes, counts)
plt.ylabel("Presence in Both Conditions")
plt.title("Shared Genes Between SARS-CoV-2 and Alzheimer's Disease")
plt.show()

```
